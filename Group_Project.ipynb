```python
products = pd.read_csv(os.path.join(path, 'olist_products_dataset.csv'))
products
```

###ตรวจสอบค่า Missing ของตาราง Order

```python
products.isnull().any()
```

```python
### function หา % ของข้อมูลที่จะหายไปถ้า .dropna()

def loss_data(data):
    im = data.shape[0]
    jm = data.dropna().shape[0]
    km = im - jm
    print(f'size before drop = {im}')
    print(f'size after drop = {jm}')
    print(f'we loss {km*100/im} % of data')
```

```python
loss_data(products)
```

```python
products_dropped = products.dropna()
products_dropped
```

##เชื่อมต่อตาราง

```python
### ต่อตาราง order_items กับ sellers

merged_ori_sellers = order_items.merge(sellers,how='left',left_on='seller_id',right_on='seller_id')
merged_ori_sellers
```

```python
### ต่อตาราง order_items, sellers, products

merged_ori_sellers_pro = merged_ori_sellers.merge(products,how='left',left_on='product_id',right_on='product_id')
merged_ori_sellers_pro
```

##ตรวจสอบค่า Missing ของตาราง ที่เชื่อมต่อกันแล้ว

```python
merged_ori_sellers_pro.isnull().any() # ดูว่าแต่ละ column มีค่า missing มั้ย #พบว่ามีค่าmissing
```

```python
loss_data(merged_ori_sellers_pro)#เช็คว่าค่าmissingที่พบสามารถยอมรับได้หรือไม่ #ยอมรับได้เพราะมีค่าแค่ 1.42%
```
